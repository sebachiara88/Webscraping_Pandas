
#WebScraping using Pandas

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2020_per_game.html'
```

## **Reading HTML webpage into pandas**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_html(url, header = 0)
df
```

## Numbers of tables in the HTML

```python
len(df)
```

## Data Cleaning

```python
df2020 = df[0]
```

```python
df2020[df2020.Age == 'Age']
```

```python
df = df2020.drop(df2020[df2020.Age == 'Age'].index) #Removing the subheaders
```

```python
df.shape
```

```python
datatype = df.dtypes
datatype
```

```python
df.head()
```

```python
df.fillna("0") #reaplacing NaN values with a zero
```

```python
df=df.drop_duplicates(subset='Player',keep='first') #As some players appears more than once because they moved from one team to another, I removed duplicates using Player column
```

```python
df.sort_values(by='Player', ascending=True)
```

```python
sns.distplot(df.PTS, kde=False) #Making an histogram with column PTS
```

##Calculating and visualize score's mean for every team

```python
team_points = df[['Tm', 'PTS']]
team_points = team_points.rename(columns={'Tm': 'Teams', 'PTS': 'AverageScoring'}).reset_index(drop=True)
team_points
```

```python
team_points['AverageScoring'] = team_points.AverageScoring.astype(float) # changing the DType on column PTS from string to numeric
```

```python
team_points.dtypes
```

```python
team_points = team_points.groupby(by = 'Teams').mean().sort_values(by='AverageScoring', ascending=False).reset_index()
```

```python
team_points
```

```python
fig = plt.gcf()
fig.set_size_inches(15, 10)
ax = sns.barplot(x="Teams", y="AverageScoring", data=team_points, palette="Blues_d")
```
